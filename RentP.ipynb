```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r'C:\Users\RYZEN-5\OneDrive\Desktop\Machine Learning\House Rent Prediction\House_Rent_Dataset.csv')
df.head()
```

```python
df.info()
```

```python
print("Mean House Rent =", round(df["Rent"].mean()))

print("Median House Rent =", round(df["Rent"].median()))

print("Highest House Rent =", round(df["Rent"].max()))

print("Lowest House Rent =", round(df["Rent"].min()))
```

```python
plt.figure(figsize = (16, 4))
plt.ticklabel_format(style = 'plain')
plt.scatter(df["Size"], df["Rent"])
plt.xlabel("Size")
plt.ylabel("Rent")
```

```python
X = df['Size']
Y = df['Rent']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0, test_size=.15)
```

```python
plt.figure(figsize = (16, 4))
plt.ticklabel_format(style = 'plain')
plt.scatter(X_train, Y_train, alpha=0.7, label='Training data', color='r')
plt.scatter(X_test, Y_test, alpha=0.7, label='Testing Data', color='g')
plt.xlabel('Size')
plt.ylabel('Rent')
plt.legend()
```

```python
linr = LinearRegression()
```

```python
linr.fit(X_train.values.reshape(-1,1), Y_train.values)
```

```python
Y_pred = linr.predict(X_test.values.reshape(-1,1))

plt.figure(figsize = (16, 4))
plt.ticklabel_format(style = 'plain')
plt.plot(X_test, Y_pred, label='Linear Regression', color='r')
plt.scatter(X_test, Y_test, label='Actual Data', color='g')
plt.xlabel('Size')
plt.ylabel('Rent')
plt.legend()
```

```python
rent = linr.predict(np.array([[1500]]))[0]
print("The Rent for a house of 1500 sq.ft is Rs.",rent,".")
```
